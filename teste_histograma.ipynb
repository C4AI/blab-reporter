```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import random

dados = pd.read_csv('rj_dados_temp.csv').dropna()
dados = dados.query('temperatura >= 12 and temperatura <= 30')
dados.columns = ['Temperatura média']
dados['Temperatura média'] = dados['Temperatura média'] * 8.5 + random.randint(1, 100)

median = dados['Temperatura média'].median()
std = dados['Temperatura média'].std()
percentile_mais = dados['Temperatura média'].quantile(0.95)
percentile_menos = dados['Temperatura média'].quantile(0.05)

dados.columns = ['Histograma da quantidade de navios no porto de Santos (SP)']

ax = sns.displot(dados['Histograma da quantidade de navios no porto de Santos (SP)'],
             bins=int(18), color = 'darkblue', 
             stat='probability')
ax.fig.set_size_inches(8,5)
plt.axvline(median + 3, color = 'Black', label = 'Median', linestyle='dashed')
plt.axvline(percentile_menos, color = 'Green', label = 'Percentile 5%', linestyle='dashed')
plt.axvline(percentile_mais, color = 'Orange', label = 'Percentile 95%', linestyle='dashed')
plt.axvline(300, color = 'Red', label = 'Critical Point', linestyle='dashed')
plt.legend()
plt.show()
```
