```python
import csv
import requests
import pandas as pd

CSV_URL = 'http://www.moho.iag.usp.br/fdsnws/event/1/query?format=text&orderby=time&limit=100&maxlat=10&minlon=-55&maxlon=-10&minlat=-55'


page = requests.Session()

download = page.get(CSV_URL)

decoded_content = download.content.decode('utf-8')

cr = csv.reader(decoded_content.splitlines(), delimiter='|')
my_list = list(cr)
lista_colunas = ['#EventID', 'Time', 'Latitude', 'Longitude', 'Depth/km', 'MagType', 'Magnitude', 'EventLocationName']
df = pd.DataFrame(my_list[1::], columns=my_list[0])[lista_colunas]
df
```
